<h2>1.Full-backs as Playmakers</h2>
<p> The term "Full-backs as Playmakers" refers to a tactical approach in football where the team's full-backs are given a significant role in the creation and distribution of attacking plays.Traditionally, full-backs were primarily defensive players responsible for marking wingers and supporting the team's defense. However, Jürgen Klopp have embraced an attacking philosophy that involves full-backs actively participating in offensive maneuvers.</p>


```python
def clean_numeric(value):
    # Define a regex pattern to match valid numeric values
    numeric_pattern = re.compile(r'^\d+$')  # Updated pattern for integers

    # Check if the value matches the numeric pattern
    return numeric_pattern.match(value)

def get_fullback_data(url, output_filename='fullback_data.csv'):
    # Send an HTTP request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, "html.parser")

        # Extract and print the data from the 5th table
        tables = soup.find_all("table")
        data_1 = tables[4]  # Index 4 corresponds to the 5th table

        # Extract and store the rows for the selected columns in a list of lists
        player_data = []
        for row in data_1.find_all("tr")[2:]:  # Skip the header row and the first data row
            columns = row.find_all(["th", "td"])
            player = columns[0].text.strip()
            squad = columns[4].text.strip()
            gca = columns[14].text.strip()
            # Check if 'GCA' is a valid numeric value
            if clean_numeric(gca):
                player_data.append([player, squad, int(gca)])  # Convert 'GCA' to integer if valid

        # Create a DataFrame from the list of lists
        columns = ['Player', 'Squad', 'GCA']
        df = pd.DataFrame(player_data, columns=columns)

        # Save DataFrame to a CSV file
        df.to_csv(output_filename, index=False)
        
        
       
        # Print the modified DataFrame without the index column
        print(df.to_string(index=False))
        print(f"DataFrame saved to {output_filename}")
   
         
    else:
        print("Failed to retrieve data. Status code:", response.status_code)

url = "https://fbref.com/en/stathead/player_comparison.cgi?request=1&sum=0&comp_type=spec&spec_comps=9&player_id1=cd1acf9d&p1yrfrom=2019-2020&player_id2=86dd77d1&p2yrfrom=2019-2020&player_id3=9c94165b&p3yrfrom=2019-2020&player_id4=f4290206&p4yrfrom=2019-2020&player_id5=75a72a99&p5yrfrom=2019-2020"
get_fullback_data(url)
```

<h2>2.Defensive Solidity</h2>
<p>The term "defensive solidity" describes a team's capacity to effectively defend and keep the opposition from scoring goals. It combines disciplined individual play, well-planned defensive structures, and tactical methods to reduce the opposition's attacking opportunities. Strong defensive players usually make their team harder to beat and give up fewer goals.</p>
<p>I would use defensive statistic values like goals conceded and clean sheet to assess Liverpool's defensive solidity against the opposition. It is simpler to evaluate the overall defensive performance of one's squad using these two data sets.</p>

```python
def get_defense_data(url):
    try:
        # Send an HTTP request to the URL
        response = requests.get(url)
        response.raise_for_status()  # Raises an HTTPError for bad responses (4xx or 5xx)
    except requests.exceptions.RequestException as e:
        print(f"Error during the request: {e}")
        return None

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, "html.parser")

        # Extract and print the data
        tables = soup.find_all("table")

        # Check if tables were found
        if tables:
            # Extract data from the first table
            data_2 = tables[0]

            # Find all rows in the table
            rows = data_2.find_all("tr")

            # Extract header (assuming it's in the first row)
            header_row = rows[0]
            headers = [header.text.strip() for header in header_row.find_all(["th", "td"])]

            # Find the index of the 'Avg per game' column
            avg_column_index = headers.index('Avg per game')

            # Remove the 'Avg per game' column and its values from headers
            headers.pop(avg_column_index)

            # Initialize a list to store the data
            defense_data = []

            # Extract data for each row (excluding the 'Avg per game' column)
            for row in rows[1:6]:  # Display the first 5 rows
                row_data = [
                    cell.text.strip() if re.match(r'^\d+(\.\d+)?$', cell.text.strip()) else cell.text.strip()
                    for index, cell in enumerate(row.find_all(["th", "td"])) if index != avg_column_index
                ]
                defense_data.append(row_data[1:])  # Adjusted here
            # Create a DataFrame using Pandas
            df1 = pd.DataFrame(defense_data, columns=headers[1:])  # Adjusted here

            # Set the index to start from 1
            df1.index = range(1, len(df1) + 1)

            # Display the DataFrame
            print(df1)

            # Save the DataFrame to a CSV file
            df1.to_csv('defense_data.csv', index_label='Index')
            print("DataFrame saved to defense_data.csv")

            return df1  # Return the DataFrame 
        else:
            print("No tables found on the page.")
            return None
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

# Example usage
url = "https://www.sportsmole.co.uk/football/premier-league/2019-20/best-defence.html"
defense_data = get_defense_data(url)
```

```python
def get_league_table(url):
    try:
        # Make a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful (status code 200)
        response.raise_for_status()

        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the table containing the league standings
        table = soup.find('table')

        # Use Pandas to read the HTML table into a DataFrame, specifying index_col=0
        league_table = pd.read_html(str(table), index_col='Team')[0]

        return league_table

    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve the page. Error: {e}")
        return None

# Example usage
url = 'https://www.sportinglife.com/football/league-tables/english-premier-league/1/season/2019-2020/40'
result = get_league_table(url)

if result is not None:
    # Save the extracted data into a DataFrame
    df3 = pd.DataFrame(result)

    # Display the DataFrame
    print(df3)

    # Optionally, you can export the DataFrame to a CSV file
    df3.to_csv('league_table.csv', index=True)
```

<p>I wanted to highlight the significant advantage Liverpool had over their nearest rival in the standings during their championship-winning campaign, since this emphasises Kloops' astute strategic thinking and planning, which were crucial to their successful campaign.</p>

```python
# Load DataFrame from the CSV file
output_filename = 'league_table.csv'  # Update with your actual filename
df3 = pd.read_csv(output_filename)
# Sort the DataFrame by 'PTS' in descending order
df3_sorted = df3.sort_values(by='PTS', ascending=False)

# Set the style using seaborn
sns.set(style="whitegrid")

# Create a bar plot with seaborn for better aesthetics
plt.figure(figsize=(14, 8))
bar_plot = sns.barplot(x=df3_sorted.index, y=df3_sorted['PTS'], palette='coolwarm', errorbar=None)
bar_plot.set(xlabel='Team', ylabel='Points (PTS)')
plt.title('Points Distribution of Premier League Teams', fontsize=15)
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate team names for better readability
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a horizontal grid for reference

# Annotate the bars with point values
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                      textcoords='offset points')

plt.tight_layout()

# Display the plot
plt.show()
```
